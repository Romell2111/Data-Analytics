```python
#loading the dataset
import pandas as pd
import numpy as np
from pandas import read_csv
df = pd.read_csv('C:\\Users\\Romell\\Anaconda3\\Reviews.csv', delimiter=',', encoding="utf-8-sig")
df.head()
```

```python
#Cleaning dataset (Finding Null values)
df.isnull().sum()
#Removing Null values
df.dropna(inplace=True)
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
f1 = df['Score'].values
f2 = df['HelpfulnessNumerator'].values
X=np.matrix(list(zip(f1,f2)))

#K-Means Clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(f1, f2, c=y_kmeans, s=200, cmap='inferno')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], s=400, alpha=0.9);
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from subprocess import check_output
labels = np.array(df)
dataset = df.iloc[1:10,5:7]
dataset.head()



Z = linkage(dataset, 'ward')
plt.figure(figsize=(50, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Dataset')
plt.ylabel('Distance')
from pylab import rcParams
rcParams['figure.figsize'] = 30, 60
dendrogram(
    Z,
    labels=labels,
    leaf_rotation=90.,
    leaf_font_size=20.,
)
plt.show()
```
